```python
import pandas as  pd
```

```python


# Use the raw file link instead of the HTML page
url = "https://raw.githubusercontent.com/eumarcelorodrigues/Projeto-de-Ingles-Abusivo/main/train.csv"

df = pd.read_csv(url)

print(df.head())  # show first 5 rows
```

```python
df.info()
```

```python
display(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contagem das classes
label_counts = df['label'].value_counts()

# Usando paleta colorblind-friendly
sns.set_palette("colorblind")

plt.figure(figsize=(6,4))
sns.barplot(x=label_counts.index, y=label_counts.values)
plt.title('Contagem de Frases Ofensivas vs Não Ofensivas')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
